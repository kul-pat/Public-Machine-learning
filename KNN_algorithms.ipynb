```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,f1_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(axis=1,inplace=True)
df
```

```python
df.set_index('id',inplace=True)
df
```

```python
df.shape
```

```python
df.loc[842302]
```

```python
# encoding the categorical data values
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
df.iloc[:,0] = labelencoder_y.fit_transform(df.iloc[:,0].values)
df.iloc[:,0]
```

```python
# split dataset
X =df.drop('diagnosis',axis =1)
y =df['diagnosis']
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
X
```

```python
y
```

```python
# scale the data (" feature sclaing")
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
X_train
```

```python
import math
math.sqrt(len(y_test))
```

```python
# Define the model : KNN
classifier =KNeighborsClassifier(algorithm='auto',weights='uniform',leaf_size=1,p=1,metric='manhattan',n_neighbors=11)
```

```python
#fit_model
classifier.fit(X_train,y_train)
```

```python
# predict the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# evluate the model
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(f1_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
##Randomized Search Cv
from sklearn.model_selection import GridSearchCV
p=[1,2]
weights=['uniform','distance']
algorithm =['auto','ball_tree','kd_tree','brute']
metric = ['euclidean','manhattan','minkowski']
# Create the random grid
random_grid = {'n_neighbors':np.arange(1,50),
               'leaf_size':np.arange(1,20),
               'p': p,'weights': weights,'algorithm':algorithm,
               'metric': metric
            }
print(random_grid)
```

```python
# using the RandomizesearchCV 
knn_2 = KNeighborsClassifier()
clf =GridSearchCV(knn_2,random_grid,cv=5,scoring='accuracy')
clf.fit(X_train,y_train)
```
